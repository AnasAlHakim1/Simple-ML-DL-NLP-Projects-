**Append Mode**

```python
file = open ('NLP.txt' , 'a+')

file.write("\nNow the file has more content! ")

file.seek(0)

print (file.read())

file.close()

```

**Write Mode**

```python
file = open ('NLP.txt' , 'w+')

file.write("Oops! I have deleted the content! ")

file.seek(0)

print (file.read())

file.close()
```

**Mini Challenge1:**

```python
file = open ('Myfile.txt' , 'w+')

file.write( "Hello there!" '\n' * 100)

file.seek(0)

for i in range(6):
   print(file.readline())

file.close()

```

**With method**

```python
with open ('Myfile.txt' , 'a+') as file:

   file.write( "I love python")

   file.seek(0)

   sum = 0

   for i in file :
      sum = sum + 1

print('The number of lines is: ' , sum)
```

**Writelines**

```python
names = ("Ahmed\n" , "Omar\n" , "Khaled")
print(type(names))
file = open ("data.txt" , "w+")

file.writelines (names[2])

file.close()
```
